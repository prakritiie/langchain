## 1. Streaming

Streaming means receiving the model response token-by-token instead of waiting for the full output.
Useful for chat apps, realtime UIs, and faster perceived response times.

```python
import os
from dotenv import load_dotenv

from langchain_groq import ChatGroq

load_dotenv()

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROQ_API_KEY")
)

response = llm.invoke("Explain AI agents simply")

print(response.content)
```

## 2. Batching

Batching means sending multiple inputs together in a single request.
It improves throughput and reduces API overhead when processing many prompts.

```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langserve import client

load_dotenv()

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROQ_API_KEY")
)

prompts = [
    "What is LangChain?",
    "What is an AI Agent?",
    "Explain RAG briefly"
]

for prompt in prompts:
    response = llm.invoke(prompt)

    print("\nPrompt:", prompt)
    print("Answer:", response.content)
```

## 3. Messages

Definition
Messages are structured chat inputs containing roles like system, user, and assistant.
They help maintain conversation history and context.

```python
import os
from dotenv import load_dotenv

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_groq import ChatGroq
from langchain_core.messages import (
    HumanMessage,
    SystemMessage
)

load_dotenv()

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROQ_API_KEY")
)

messages = [
    SystemMessage(content="You are a helpful AI tutor."),
    HumanMessage(content="Explain middleware in AI agents")
]

response = llm.invoke(messages)

print(response.content)
```

## 4. Data Classes

Definition
Dataclasses are Python classes used for structured data storage.
In AI agents, they help maintain clean state, memory, and workflow data.

```python
from dataclasses import dataclass

@dataclass
class AgentState:
    user_query: str
    response: str
    tokens_used: int

state = AgentState(
    user_query="What is LangChain?",
    response="LangChain is an AI framework.",
    tokens_used=120
)

print(state)
```

## 5. Middleware

Middleware acts as an intermediate processing layer between user input and model execution.
It is commonly used for logging, validation, caching, and modifying requests/responses.

```python
def logging_middleware(func):

    def wrapper(user_input):
        print("User Input:", user_input)

        result = func(user_input)

        print("Model Output:", result)

        return result

    return wrapper


@logging_middleware
def ai_agent(user_input):
    return f"AI says: {user_input}"


ai_agent("Explain streaming")
```

## 5.1 Summarization Middleware

Summarization middleware compresses long conversation history into shorter summaries.
This helps reduce token usage and maintain context efficiently.

```python
import os
from dotenv import load_dotenv

from langchain_groq import ChatGroq

load_dotenv()

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROQ_API_KEY")
)

conversation = """
User: Explain AI agents
Assistant: AI agents are autonomous systems...

User: Explain tools in agents
Assistant: Tools allow agents to interact externally...
"""

summary_prompt = f"""
Summarize this conversation in 3 lines:

{conversation}
"""

response = llm.invoke(summary_prompt)

print(response.content)
```

## 5.2 Human in the Feedback Loop

Definition
Human-in-the-loop (HITL) systems allow humans to review or approve AI outputs before final execution.
This improves reliability, safety, and decision quality in AI agents.

```python
def generate_response(query):
    return f"Generated answer for: {query}"


query = "Delete all database records"

ai_response = generate_response(query)

print("AI Response:", ai_response)

approval = input("Approve action? (yes/no): ")

if approval.lower() == "yes":
    print("Action Executed")
else:
    print("Action Cancelled")
```
