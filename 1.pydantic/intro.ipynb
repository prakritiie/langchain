```python
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    id: int
    name: str
    city: str

person = Person(id=1, name="John Doe", city="New York")

print(person)
```

```python
type(person)
```

```python
person1 = Person(id=2, name="Jane Doe", city=12)

## we will get an error because city is expected to be a string but we are passing an integer
```

```python
from dataclasses import dataclass

@dataclass
class PersonDataClass:
    id: int
    name: str
    city: str

person_dataclass = PersonDataClass(id=1, name="Prachi", city="India")
print(person_dataclass)
```

```python
person2 = PersonDataClass(id=2, name="Jane Doe", city=12)

## we will not get an error because dataclass does not perform any type checking
```

## 2. Model with Optional Field 

add optional fields using Python's optional type:

```python
from typing import Optional
class Employee(BaseModel):
    id: int
    name: str
    department: Optional[str] = None
    salary: Optional[float] = None # default value is None
    is_manager: Optional[bool] = True # default value is True


employee1 = Employee(id=1, name="Alice", department="HR", salary=50000.0, is_manager=True)
print(employee1)
```

```python
employee2 = Employee(id=2, name="Bob")
print(employee2)
```

```python
from pydantic import BaseModel
from typing import List

class Classroom(BaseModel):
    room_number : str
    students : List[str]   # List of strings representing student names
    capacity : int
```

```python
#create a classroom instance
classroom = Classroom(room_number="101A",
                       students=["Alice", "Bob", "Charlie"], 
                       capacity=30)

print(classroom)
```

```python
try:
    classroom_invalid = Classroom(room_number="102B", students="Alice,123", capacity=30)
except Exception as e:
    print("Error creating classroom instance:", e)  

```

## Models with nested structures

```python
from pydantic import BaseModel

class Address(BaseModel):
    street: str
    city: str
    zip_code: str

class Customer(BaseModel):
    id: int
    name: str
    email: str
    address: Address  # Nested model

# Create an instance of the nested model
customer = Customer(id=1, name="John Doe", email="john.doe@example.com", address=Address(street="123 Main St", city="Anytown", zip_code="12345"))

print(customer)

```

### pydantic fields: customizable and constraints

The feild function in Pydantic enhances model fields beyond basic type hints by allowing you to sepcify validation rules , default values, aliases , and  more. 

```python
from pydantic import BaseModel, Field
class Item(BaseModel):
    name:str = Field(min_length = 2, max_length=50)  # name must be a non-empty string
    price: float = Field(gt=0, le = 1000)         # price must be greater than 0    
    quantity: int = Field(ge=1, le=100)           # quantity must be between 1 and 100
    
```

```python
item = Item(name="Laptop", price=999.99, quantity=10)
print(item)
```
